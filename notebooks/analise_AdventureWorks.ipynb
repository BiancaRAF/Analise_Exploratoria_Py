```python
#Criando o dataframe
df = pd.read_excel("AdventureWorks.xlsx")

df.sample(5)
```

```python
#Verificar a quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos das colunas
df.dtypes
```

```python
#Qual a receita total?

df["Valor Venda"].sum()
```

```python
#Qual o custo total?

df["Custo Total"] = df["Custo Unitário"].mul(df["Quantidade"])

df.sample(4)
```

```python
round(df["Custo Total"].sum(),2)
```

```python
#Qual o lucro total?

df["Lucro"] = df["Valor Venda"] - df["Custo Total"]

df.sample(4)
```

```python
df["Lucro"].sum()
```

```python
df.dtypes
```

```python
#Quanto tempo demorou o envio?

df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"]

df.sample(3)
```

```python
#Média do tempo de envio para cada marca
  #Como na coluna temos a informação "days" é necessário trazer apenas a quantidade de dias. Se verificarmos o tipo da coluna percebemos que é diferente do datetime64
df.dtypes
```

```python
#Extraindo apenas os dias
df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days

#Apesar de quê fazendo o comando df.groupby("Marca")["Tempo Envio"].mean() para o tipo de dado anterior retornou a média normalmente dos dados

df.sample(3)
```

```python
df["Tempo Envio"].dtype
```

```python
df.groupby("Marca")["Tempo Envio"].mean()
```

```python
#Verificando se tem dados faltantes
df.isnull().sum()

df.columns
```

```python
#Qual o lucro por ano e marca?

df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#Retirar o formato científico

pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Trazer em formato de tabela - dataframe | Só colocar o método .reset_index()

lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()

lucro_ano
```

```python
#Qual o valor total de produtos vendidos?

produto_vendido = df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)

produto_vendido
```

```python
#Gráfico Total de produtos vendidos

produto_vendido.sort_values(ascending=True).plot.barh(title = "Total de produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#Lucro por ano

lucro_ano = df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()

lucro_ano.plot.bar(title = "Lucro por ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
lucro_ano
```

```python
#Criando um novo dataframe apenas com os dados de 2009

df_2009 = df[df["Data Venda"].dt.year == 2009]

df_2009.sample(3)
```

```python
#Gráfico de linhas com lucro por mês

lucro_mes_2009 = df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum()

lucro_mes_2009
```

```python
lucro_mes_2009.plot(title = "Lucro por mês - 2009")
plt.xlabel("Mês")
plt.xlabel("Lucro");
```

```python
#Lucro por Marca

lucro_marca_2009 = df_2009.groupby("Marca")["Lucro"].sum()

lucro_marca_2009
```

```python
lucro_marca_2009.plot.bar(title = 'Lucro por marca - 2009')
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal"); #Colocado o rótulo do eixo x na horizontal
```

```python
#Lucro por classe

lucro_classe_2009 = df_2009.groupby("Classe")["Lucro"].sum()
lucro_classe_2009
```

```python
lucro_classe_2009.plot.bar(title = 'Lucro por classe - 2009')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation = 'horizontal');
```
